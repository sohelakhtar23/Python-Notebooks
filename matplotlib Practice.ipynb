```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([2,4,6,8,10])
y=np.array([25,4,7,45,30])
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title("Simple Plot")
ax.set_xlabel("x label")
ax.set_ylabel("y label")
plt.show()
```

```python
fig, ax2 = plt.subplots()
ax2.plot(x, y, label="discrete y")
ax2.plot(x, x**2, label="x^2")
ax2.plot(y, y**0.5, label="y^0.5")
ax2.legend()
ax2.set(title="Three plots in 1", xlabel="x label", ylabel="y label")
plt.show()
```

## Scatter plot

```python
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.scatter(x, x**2)
ax.grid()
```

### plot() string formatting

fmt = '[marker][line][color]'

```python
plt.plot(x, y, 'r+--')
```

## Bar chart

```python
company = ['Google', 'Amzn', 'Tesla', 'AliBaba']
revenue = [250, 200, 290, 110]
profit = [50, 20, 80, 20]
```

```python
xpos = np.arange(len(company))

fig, ax = plt.subplots()

ax.bar(xpos-0.2, revenue, width=0.4, label="revenue")
ax.bar(xpos+0.2, profit, width=0.4, label="profit")
ax.set(title="Tech Stocks", ylabel="revenue(billion)")
ax.set_xticks(xpos, company)
ax.legend()
plt.show()
```

## Histograms

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.8, bins=3)
```

## Pie chart

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.pie(exp_vals, labels= exp_labels)
plt.show()
```

## saving chart

```python
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("piechart.jpg", bbox_inches="tight", pad_inches=1, transparent=True)
```
