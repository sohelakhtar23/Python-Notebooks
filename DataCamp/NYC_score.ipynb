```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('NYC_schools_score.csv')
print(df.shape)
df.head()
```

```python
best_math_schools = df[df['average_math']>=640][['school_name', 'average_math']]
best_math_schools = best_math_schools.sort_values('average_math', ascending=False)
best_math_schools
```

```python
df['total_SAT'] = df['average_math'] + df['average_reading'] + df['average_writing']
top_10_schools = df.sort_values('total_SAT', ascending=False)
top_10_schools = top_10_schools[['school_name', 'total_SAT']].head(10)
top_10_schools
```

```python
temp_3 = df.groupby('borough')['total_SAT'].agg(['count', 'mean', 'std']).round(2)
largest_std_dev = temp_3[temp_3['std'] == temp_3['std'].max()]
largest_std_dev
```
